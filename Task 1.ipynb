```python
from PIL import Image
```

```python
a=Image.open('people.jpg')
```

```python
width,height=a.size
```

```python
width
```

```python
height
```

```python
#changing size
b=a.resize((int(width/8),int(height/7)))
```

```python
b
```

```python
b.save('rose.png')
```

```python
b.size
```

```python
#flip
b.transpose(Image.FLIP_LEFT_RIGHT).save('rose.png')
```

```python
#upside down
b.transpose(Image.FLIP_TOP_BOTTOM).save('people.png')
```

```python
#getting RGB values
colors=a.getpixel((240,320))
print(colors)
```

```python
#converting into greyscale
c=Image.open('rose.png').convert('LA')
c.save('rosegrey.png')
```
